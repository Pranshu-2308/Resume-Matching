```python
job_description = docx2txt.process('/content/Job desp.docx')
resume = docx2txt.process('/content/PranshuChoudhary_Resume copy.docx')
resume1 = docx2txt.process('/content/Matched Resume.docx')
```

```python
print(job_description)
```

```python
from docx import Document
import re

def clean_text(text):
     # Convert to lowercase
    text = text.lower()
    # Remove punctuation and stopwords
    text = re.sub(r'[^\w\s]', '', text)
    stopwords = set(['the', 'and', 'a', 'is', 'in', 'on', 'it', 'for', 'with', 'at', 'to', 'of', 'this', 'an'])
    words = text.split()
    words = [word for word in words if word not in stopwords]
    return ' '.join(words)



# Call the clean_text function
cleaned_job_description = clean_text(job_description)
cleaned_resume = clean_text(resume)
cleaned_resume1 = clean_text(resume1)
# Print the cleaned text
print(cleaned_job_description)
print("\n" * 3)
print(cleaned_resume)
print("\n" * 3)
print(cleaned_resume1)

```

```python
content = [job_description, resume,cleaned_resume1 ]
from sklearn.feature_extraction.text import CountVectorizer
cv = CountVectorizer()
count_matrix = cv.fit_transform(content)

from sklearn.metrics.pairwise import cosine_similarity
mat = cosine_similarity(count_matrix)
print(mat)
```

```python
print('Resume 1  Matches by: {:.0f}%'.format(mat[1][0] * 100))
print('Resume 2  Matches by: {:.0f}%'.format(mat[2][0] * 100))
print("\n" * 2)

if mat[1][0] > mat[2][0]:
    print("Resume 1 is a better fit.")
else:
    print("Resume 2 is a better fit.")
    print("\n" * 3)
```

```python
from docx import Document
from wordcloud import WordCloud
import matplotlib.pyplot as plt

def generate_and_show_wordcloud(text):
    # Generate a word cloud
    wordcloud = WordCloud(width=800, height=400, background_color='white').generate(text)

    # Plot the WordCloud image
    plt.figure(figsize=(10, 5))
    plt.imshow(wordcloud, interpolation='bilinear')
    plt.axis('off')
    plt.show()

# usage:
generate_and_show_wordcloud(cleaned_resume)
print("\n" * 3)
generate_and_show_wordcloud(cleaned_resume1)
print("\n" * 3)
generate_and_show_wordcloud(cleaned_job_description)

```
